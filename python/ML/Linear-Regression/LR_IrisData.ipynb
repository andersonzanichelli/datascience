## LR Iris Data

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dframe = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
target = pd.DataFrame(iris.target, columns=["TARGET"])
```

```python
target.head()
```

```python
df = dframe.join(target)
```

```python
df.head()
```

```python
sns.pairplot(df, hue='TARGET')
```

```python
sns.boxplot( x='petal width (cm)', y='sepal width (cm)', data=df, hue='TARGET')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('TARGET', axis=1)
```

```python
X.head()
```

```python
y = target
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ml = LinearRegression()
```

```python
ml.fit(X_train, y_train)
```

```python
predictions = ml.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
sns.distplot((y_test-predictions));
```

```python
coeffecients = pd.DataFrame(ml.coef_,['Coefficients'])
coeffecients.columns = X.columns
coeffecients
```

## Coefficients
- Holding all other features fixed, a 1 unit increase in **Sepal length** is associated with an **decrease of -0.07 cm**.
- Holding all other features fixed, a 1 unit increase in **Sepal width** is associated with an **decrease of -0.08 cm**.
- Holding all other features fixed, a 1 unit increase in **Petal length** is associated with an **increase of 0.21 cm**.
- Holding all other features fixed, a 1 unit increase in **Petal width** is associated with an **increase of 0.59 cm**.
