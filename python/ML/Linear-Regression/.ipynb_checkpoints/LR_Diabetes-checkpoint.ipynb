## Diabetes Dataset

```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes.DESCR
```

```python
dfdiab = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
```

```python
dfdiab.head()
```

```python
dftarget = pd.DataFrame(diabetes.target, columns=['TARGET'])
```

```python
dftarget.head()
```

```python
df = dfdiab.join(dftarget)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df['age'], bins=10)
```
