# Piechart

```r
#?pie
slice = c(40, 20, 40)
countries = paste(c('Brazil', 'Argentina', 'Chile'), slice)
countries = paste(countries, '%', sep = '')
countries
```

```r
pie(slice, labels = countries, col = c('darksalmon', 'gainsboro', 'lemonchiffon4'))
```

```r
values = table(Species)
labels = paste(names(values))
pie(values, labels = labels, main = 'Species Distribution')
```

```r
pie3D(slice, labels = countries, explode = 0.03, col = c('steelblue1', 'tomato2', 'tan3'), main = 'Sales Distribution')
```
