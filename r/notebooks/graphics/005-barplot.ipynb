# Barplot

```r
#?barplot
data = matrix(c(653, 1342, 34, 734, 34, 23, 219, 88, 34, 234, 12,19), nrow = 3, byrow = T)
data
```

```r
colnames(data) = c('0', '1-150', '151-300', '> 300')
rownames(data) = c('Young', 'Adult', 'Old')
data
```

```r
barplot(data, beside = T)
```

```r
#Stackedbar
barplot(data)
```

```r
barplot(data, beside = T, col = c('aquamarine', 'tan3', 'seagreen3'))
legend("topright", pch = 1, c('Young', 'Adult', 'Old'), col = c('aquamarine', 'tan3', 'seagreen3'))
```

```r
barplot(t(data), beside = T, col = c('aquamarine', 'tan3', 'seagreen3', 'aliceblue'))
legend("topright", pch = 1, c('0', '1-150', '151-300', '> 300'), col = c('aquamarine', 'tan3', 'seagreen3', 'aliceblue'))
```
