# Maps in R

```r
install.packages(c('ggplot2', 'maps', 'mapdata'))
```

```r
library(ggplot2)
library(maps)
library(mapdata)
```

```r
map = map_data('world')
```

```r
head(map)
```

```r
ggplot() +
geom_polygon(data = map, aes(x=long, y = lat, group = group),
            fill = 'seagreen1', color = 'black') +
coord_fixed(1.3)
```

```r
ggplot() +
geom_polygon(data = map, aes(x=long, y = lat, fill = region, group = group)) +
coord_fixed(1.3) +
guides(fill = F)
```
