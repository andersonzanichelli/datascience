# Histogram

```r
data = c(18, 20, 20, 21, 22, 24, 25, 25, 26, 27, 29, 29,
         30, 30, 31, 31, 32, 33, 34, 35, 36, 36, 37, 37,
         37, 37, 38, 38, 38, 40, 41, 43, 44, 44, 45, 45,
         45, 46, 47, 48, 49, 50, 51, 53, 54, 54, 56, 58, 62, 65)
```

```r
hist(data, main = 'Number of daily accidents', xlab = 'Accidents', ylab = 'Frequency')
```

```r
hist(data, main = 'Number of daily accidents', xlab = 'Accidents', ylab = 'Frequency', breaks = 6)
```

# Skewness (coefficient of asymmetry) (tail)

```r
mean(data)
sd(data)
median(data)
```

```r
install.packages('moments')
library(moments)
```

```r
#SK ~ 0 symetric data
#SK < 0 asymetric data, left tail
#SK > 0 asymetric data, right tail

skewness(data)
```

```r
# Another example
set.seed(1234)
x = rnorm(1000)
hist(x)
skewness(x)
```

# Coefficient of Kurtosis (curve flattening)

```r
# CK ~ 0 Normal distribution
# CK > 0 short tail, -> flat curve <-
# CK < 0 heavy tail, v squeezed graphic v

kurtosis(data)
```

```r
sample = rnorm(n = 10000, mean = 55, sd = 4.5)
```

```r
skewness(sample)
kurtosis(sample)
```

```r
library(ggplot2)
datasim = data.frame(sample)
ggplot(datasim, aes(x = sample), binwidth = 2) +
    geom_histogram(aes(y = ..density..), fill = 'red', alpha = 0.5) +
    geom_density(colour = 'blue') +
    xlab(expression(bold('Data'))) +
    ylab(expression(bold('Density')))
```
