# Boxplot

```r
#?boxplot
#?sleep
attach(sleep)
```

```r
# extra ~ group (relationship)
sleepboxplot = boxplot(data = sleep, extra ~ group,
                      main = 'Sleep Duration',
                      col.main = 'red', ylab = 'Hours', xlab = 'Drug')

means = by(extra, group, mean)

points(means, col = 'red')
```

```r
horizontalbox = boxplot(data = sleep, extra ~ group,
                      main = 'Sleep Duration',
                      col.main = 'red', ylab = 'Hours', xlab = 'Drug', horizontal = T)
```

```r
boxplot(data = sleep, extra ~ group,
                      main = 'Sleep Duration',
                      col.main = 'red', ylab = 'Hours', xlab = 'Drug', horizontal = T, col = c('blue', 'red'))
```
