```r
speed = cars$speed
```

```r
# frequency counting
hist(speed)
```

```r
hist(speed, breaks = 10)
```

```r
hist(speed, labels = T, breaks = c(0,5,10,20,30))
```

```r
graphic = hist(speed, labels = T, ylim = c(0, 10), breaks = 10)
xaxis = seq(min(speed), max(speed), length = 10)
yaxis = dnorm(xaxis, mean = mean(speed), sd = sd(speed))
yaxis = yaxis * diff(graphic$mids) * length(speed)

lines(xaxis, yaxis, col = 'red')
```
