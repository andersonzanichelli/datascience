```r
#png('speed-dist.png', width = 500, height = 500, res = 72)

par(mar = c(4, 3, 3, 3), col.axis = 'black')
plot(cars$speed, type = 's', col = 'red', bty = 'n', xlab = 'Cars ID', ylab = '')
text(8, 14, 'speed', cex = 0.85, col = 'red')
par(new = T)

plot(cars$dist, type = 's', bty = 'n', ann = F, axes = F, col = 'darkblue')
axis(side = 4)
text(37, 18, 'distance', cex = 0.85, col = 'darkblue')

title(main = 'Speed X Distance')

#dev.off()
```

```r
#png('bubble-chart.png', width = 500, height = 500, res = 72)

mycex = function(var, r, f = sqrt) {
    x = f(var)
    x_scaled = (x - min(x)) / (max(x) - min(x))
    r[1] + x_scaled * (r[2] - r[1])
}

plot(expectativa ~ pibpercap, data = df_1982, log = 'x', cex = mycex(pop, r = c(0.2, 10)), col = mycol[continente])
#dev.off()
```
