# Scatterplot

```r
set.seed(67)
x = rnorm(10, 5, 7)
y = rpois(10, 7)
z = rnorm(10, 6, 7)
t = rpois(10, 9)
```

```r
plot(x, y, col = 123, pch = 10, main = 'Multi Scatterplot', 
    col.main = 'red', cex.main = 1.5, xlab = 'Independent variable', ylab = 'Dependent variable')
points(z, t, col = 'blue', pch = 4)
points(y, t, col = 777, pch = 9)
legend(-6, 5.9, legend = c('Level 1', 'Level 2', 'Level 3'), col = c(123, 'blue', 777),
      pch = c(10, 4, 9), cex = 0.65, bty = 'n')
```
